```python
# checking the percentage of the each value in the weather column
countdrizzle = len(seattle[seattle['weather']=='drizzle'])
countrain = len(seattle[seattle['weather']=='rain'])
countsun = len(seattle[seattle['weather']=='sun'])
countsnow = len(seattle[seattle['weather']=='snow'])
countfog = len(seattle[seattle['weather']=='fog'])

print('percentage of drizzle: {:2f}%'.format((countdrizzle/(len(seattle['weather']))*100)))
print('percentage of rain: {:2f}%'.format((countrain/(len(seattle['weather']))*100)))
print('percentage of sun: {:2f}%'.format((countsun/(len(seattle['weather']))*100)))
print('percentage of snow: {:2f}%'.format((countsnow/(len(seattle['weather']))*100)))
print('percentage of fog: {:2f}%'.format((countfog/(len(seattle['weather']))*100)))
```

```python
# checking the distribution of the numerical column
fig,axs= plt.subplots(2,2,figsize=(10,8))
sns.histplot(data=seattle, x=seattle['precipitation'], kde=True,  ax= axs[0, 0], color = 'green')
sns.histplot(data=seattle, x=seattle['temp_max'], kde=True,  ax= axs[0, 1], color = 'yellow')
sns.histplot(data=seattle, x=seattle['temp_min'], kde=True,  ax= axs[1, 0], color = 'red')
sns.histplot(data=seattle, x=seattle['wind'], kde=True,  ax= axs[1, 1], color = 'blue')
plt.show()
```

```python
# checking out for the outlier by using boxplot

fig,axs= plt.subplots(2,2,figsize=(10,8))
sns.boxplot(data=seattle, x=seattle['precipitation'],  ax= axs[0, 0], color = 'green')
sns.boxplot(data=seattle, x=seattle['temp_max'],  ax= axs[0, 1], color = 'yellow')
sns.boxplot(data=seattle, x=seattle['temp_min'], ax= axs[1, 0], color = 'red')
sns.boxplot(data=seattle, x=seattle['wind'],   ax= axs[1, 1], color = 'blue')
plt.show()
```

```python
# removing the outliner in the columns
# Calculate IQR bounds for the columns
Q1 = seattle[['precipitation', 'wind']].quantile(0.25)
Q3 = seattle[['precipitation', 'wind']].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to retain only rows within the bounds for both columns
final_data = seattle[
    (seattle['precipitation'] >= lower_bound['precipitation']) & 
    (seattle['precipitation'] <= upper_bound['precipitation']) & 
    (seattle['wind'] >= lower_bound['wind']) & 
    (seattle['wind'] <= upper_bound['wind'])
]
```

```python
# checking back for the outlier by using boxplot after removing outlier

fig,axs= plt.subplots(2,2,figsize=(10,8))
sns.boxplot(data=final_data, x=final_data['precipitation'],  ax= axs[0, 0], color = 'green')
sns.boxplot(data=final_data, x=final_data['temp_max'],  ax= axs[0, 1], color = 'yellow')
sns.boxplot(data=final_data, x=final_data['temp_min'], ax= axs[1, 0], color = 'red')
sns.boxplot(data=final_data, x=final_data['wind'],   ax= axs[1, 1], color = 'blue')
plt.show()
```

```python
# checking the distribution of the numerical column after removing outlier
fig,axs= plt.subplots(2,2,figsize=(10,8))
sns.histplot(data=final_data, x=final_data['precipitation'], kde=True,  ax= axs[0, 0], color = 'green')
sns.histplot(data=final_data, x=final_data['temp_max'], kde=True,  ax= axs[0, 1], color = 'yellow')
sns.histplot(data=final_data, x=final_data['temp_min'], kde=True,  ax= axs[1, 0], color = 'red')
sns.histplot(data=final_data, x=final_data['wind'], kde=True,  ax= axs[1, 1], color = 'blue')
plt.show()
```

```python
# encoding the weather categorical into numerical
lc =LabelEncoder()
final_data['weather'] =lc.fit_transform(final_data['weather'])
```
